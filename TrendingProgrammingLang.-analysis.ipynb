```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("QueryResults.csv")
data.columns = ["DATE" , "TAG", "POSTS"]
data.head()

```

```python
data.tail()
```

```python
data.shape
```

```python
# conting the number of enteries in each column
#removing the unsupported the data format and observing the changes
data.dropna()
data.shape
```

```python
data.count
```

```python
data.groupby("TAG").sum()
```

```python
data.groupby("TAG").count()
```

```python
data.DATE = pd.to_datetime(data.DATE)
data.head()
```

```python
arr_data = data.pivot(index = "DATE" , columns = "TAG" , values = "POSTS")
```

```python
arr_data.head()
```

```python
arr_data.shape
```

```python
arr_data.count()
```

```python
arr_data.fillna(0, inplace= True)
arr_data.head()
```

```python
arr_data.isna().values.any()
```

```python
plt.plot(arr_data.index , arr_data["java"])
plt.plot(arr_data.index , arr_data["python"])
```

```python
for columnq in arr_data.columns:
  plt.plot(arr_data.index , arr_data[columnq])
```

```python

plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Posts', fontsize=14)
plt.ylim(0, 35000)
 
for column in arr_data.columns:
    plt.plot(arr_data.index, arr_data[column], 
             linewidth=3, label=arr_data[column].name)
 
plt.legend(fontsize=16)
```

```python
    # The window is number of observations that are averaged
roll_df = arr_data.rolling(window=2).mean()

plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Posts', fontsize=14)
plt.ylim(0, 35000)
     
    # plot the roll_df instead
for column in roll_df.columns:
  plt.plot(roll_df.index, roll_df[column], 
    linewidth=3, label=roll_df[column].name)
     
plt.legend(fontsize=16)
```
